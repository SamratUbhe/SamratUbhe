```python
# 1. Define a class Car with attributes and a method to display details. [cite: 46]
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
    def display_details(self):
        print(f"Car: {self.brand} {self.model}")
my_car = Car("Toyota", "Corolla")
my_car.display_details()
```

```python
# 2. Explain the purpose of the __init__ method. [cite: 47]
class User:
    def __init__(self, username):
        self.username = username
user1 = User("Aditya")
print(user1.username)
```

```python
# 3. Create a subclass ElectricCar that inherits from Car. [cite: 48]
class ElectricCar(Car):
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity
my_ev = ElectricCar("Tesla", "Model 3", "75 kWh")
my_ev.display_details()
```

```python
# 4. Demonstrate "Method Overriding" in inheritance. [cite: 49]
class Animal:
    def make_sound(self):
        print("Some generic sound")
class Dog(Animal):
    def make_sound(self):
        print("Bark")
my_dog = Dog()
my_dog.make_sound()
```

```python
# 5. Explain "Encapsulation" by creating private attributes using __ [cite: 50]
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
    def get_balance(self):
        return self.__balance
account = BankAccount(1000)
print(account.get_balance())
```

```python
# 6. Use @property to create getter and setter methods. [cite: 51]
class Employee:
    def __init__(self, salary):
        self._salary = salary
    @property
    def salary(self):
        return self._salary
    @salary.setter
    def salary(self, value):
        if value > 0:
            self._salary = value
emp = Employee(5000)
emp.salary = 6000
print(emp.salary)
```

```python
# 8. Create a class method using the @classmethod decorator. [cite: 53]
class School:
    name = "Sahyadri national school"
    @classmethod
    def change_school_name(cls, new_name):
        cls.name = new_name
School.change_school_name("Unlox Academy")
print(School.name)
```

```python
# 9. Explain the difference between a class method and a static method. [cite: 54]
class MathOperations:
    @classmethod
    def info(cls):
        return f"This is class {cls.__name__}"
    @staticmethod
    def add(a, b):
        return a + b
print(MathOperations.info())
print(MathOperations.add(5, 10))
```

```python
# 10. Write a try-except-finally block to handle a ZeroDivisionError. [cite: 55]
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
finally:
    print("Execution complete.")
```
